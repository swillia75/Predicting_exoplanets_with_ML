```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
# Read csv to dataframe
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#Get list of columns (features)
df.columns
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_disposition', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_time0bk',  'koi_impact', 'koi_duration', 
        'koi_depth',  'koi_prad', 'koi_teq', 'koi_insol', 
        'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']]
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
#Select features for model

X = selected_features.drop("koi_disposition", axis=1)
y = selected_features["koi_disposition"]

print(X.shape, y.shape)

```

```python
#Import train_test_split and split to training / test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# View X_train

X_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale data using MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)


X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Import RandomForestClassifier and train

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
```

```python
#Get Score for RandomForestClassifier

rf.score(X_test_scaled, y_test)
```

```python
#Perform feature_importance
#Removing lower importance features had no impact on accuracy
feature_importance = pd.Series(rf.feature_importances_).sort_values(ascending=False)
feature_importance
```

# Train the Model



```python
#Print data scores for training aand test datasets
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {"bootstrap": [True],
              "max_depth": [80, 90, 100, 110],
#              "max_features": [2,3],
 #            "min_samples_leaf": [3, 4, 5],
              "n_estimators": [100, 200]
              }
grid = GridSearchCV(rf, param_grid, verbose=3)
grid



```

```python
# Train the model with GridSearch

grid.fit(X_train_scaled, y_train)
```

```python
#Print best parameter and score 

print(grid.best_params_)
print(grid.best_score_)
```

```python
#Make prediction using test data

predictions = grid.predict(X_test_scaled)
```

```python
#Print accuracy score for test data

print('Test Acc: %.3f' % grid.score(X_test_scaled, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
